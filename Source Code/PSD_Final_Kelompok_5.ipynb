##### Bandpass Filter Farel

```python
# BANDPASS FILTER
import librosa
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.signal import butter, filtfilt

DATASET_PATH = "/content/drive/Shareddrives/Pengolahan Sinyal Data/Dataset/Farel"
files = [f"farel{i}.wav" for i in range(1, 11)]

# Band-pass
def bandpass_filter(y, sr, low=80, high=1100):
    nyquist = sr / 2
    b, a = butter(4, [low/nyquist, high/nyquist], btype='band')
    return filtfilt(b, a, y)

plt.figure(figsize=(12,5))
for filename in files:
    path = os.path.join(DATASET_PATH, filename)
    y, sr = librosa.load(path, sr=16000)

    # Terapkan filter
    y_filt = bandpass_filter(y, sr)

    fft_f = np.abs(np.fft.rfft(y_filt))
    freqs_f = np.fft.rfftfreq(len(y_filt), 1/sr)

    plt.plot(freqs_f, fft_f, label=filename)

plt.xlim(0, 5000)
plt.title("Perbandingan FFT Farel - After Filter")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitude")
plt.legend(fontsize="small")
plt.grid(True)
plt.show()
```

##Rama, Kafka, Farel

##### Before Filter

Pada kode ini digunakan untuk membandingkan spektrum frekuensi suara dari beberapa pembicara. Data suara dikelompokkan berdasarkan folder (Rama, Kafka, dan Farel), di mana setiap folder merepresentasikan satu pembicara. Setiap file audio dibaca dengan sampling rate asli, kemudian disamakan durasinya menjadi 3 detik agar hasil FFT dapat dibandingkan secara konsisten. Sinyal suara selanjutnya ditransformasikan ke domain frekuensi menggunakan Fast Fourier Transform (FFT) untuk memperoleh nilai magnitudo spektrum. Magnitudo FFT dari seluruh file dalam satu folder dirata-ratakan sehingga diperoleh spektrum frekuensi rata-rata yang merepresentasikan karakteristik suara pembicara. Hasil akhir ditampilkan dalam bentuk grafik untuk memperlihatkan perbedaan spektrum antar folder.

Visualisasi:

Grafik ini menampilkan spektrum frekuensi rata-rata hasil FFT dari setiap folder pembicara (Rama, Kafka, dan Farel). Setiap kurva merupakan hasil perataan magnitudo FFT dari seluruh file audio dalam satu folder sehingga mewakili karakteristik frekuensi umum masing-masing pembicara. Sumbu frekuensi dibatasi hingga 5000 Hz karena berada pada rentang dominan suara manusia. Berdasarkan visualisasi, pembicara Rama memiliki puncak magnitudo tertinggi dengan nilai sekitar 780, diikuti oleh Farel dengan puncak sekitar 750, sedangkan Kafka menunjukkan puncak yang lebih rendah, yaitu sekitar 200. Perbedaan pola dan puncak spektrum ini menunjukkan adanya perbedaan karakteristik suara antar pembicara.

```python
import os
import numpy as np
import librosa
import matplotlib.pyplot as plt

base_path = '/content/drive/Shared drives/Pengolahan Sinyal Data/Dataset/'
folders = ["Rama", "Kafka", "Farel"]

def compute_fft(wav_path):
    y, sr = librosa.load(wav_path, sr=None)

    # Normalisasi panjang supaya adil
    y = librosa.util.fix_length(y, size=sr * 3)  # paksa 3 detik (contoh)

    # FFT
    Y = np.fft.rfft(y)
    freq = np.fft.rfftfreq(len(y), 1/sr)
    magnitude = np.abs(Y)
    return freq, magnitude

# Simpan spektrum rata-rata per folder
folder_spectra = {}

for folder in folders:
    spectra = []
    full_path = os.path.join(base_path, folder)

    for file in os.listdir(full_path):
        if file.endswith(".wav"):
            path = os.path.join(full_path, file)
            freq, mag = compute_fft(path)
            spectra.append(mag)

    # rata-rata spektrum
    spectra = np.array(spectra)
    avg_spectrum = spectra.mean(axis=0)

    folder_spectra[folder] = (freq, avg_spectrum)

# Plot untuk perbandingan
plt.figure(figsize=(12, 5))
for folder, (freq, avg_spectrum) in folder_spectra.items():
    plt.plot(freq, avg_spectrum, label=folder)

plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.title("Perbandingan Spektrum Rata-rata per Folder")
plt.legend()
plt.xlim(0, 5000)
plt.show()
```

##### After Filter

Pada implementasi ini, sinyal suara dianalisis setelah melalui proses band-pass filtering. Setiap file audio dibaca dengan sampling rate asli dan diberagamkan durasinya menjadi 3 detik agar hasil analisis konsisten. Sinyal kemudian difilter pada rentang frekuensi 80–1100 Hz untuk mempertahankan komponen utama suara manusia dan mengurangi noise. Selanjutnya, sinyal ditransformasikan ke domain frekuensi menggunakan FFT (Fast Fourier Transform) dan nilai magnitudonya dirata-ratakan untuk setiap folder pembicara. Hasil spektrum rata-rata setelah filtering ditampilkan dalam grafik dengan batas frekuensi hingga 5000 Hz untuk membandingkan karakteristik suara antar pembicara.

Visualisasi:

Grafik ini menampilkan spektrum frekuensi rata-rata hasil FFT dari setiap folder pembicara (Rama, Kafka, dan Farel) setelah proses band-pass filtering. Setiap kurva merupakan hasil perataan magnitudo FFT sehingga mewakili karakteristik frekuensi umum masing-masing pembicara. Sumbu frekuensi dibatasi hingga 5000 Hz karena berada pada rentang dominan suara manusia. Berdasarkan visualisasi, Rama memiliki puncak magnitudo tertinggi sekitar 780, diikuti Farel sekitar 750, sedangkan Kafka memiliki puncak lebih rendah sekitar 200, yang menunjukkan perbedaan karakteristik suara antar pembicara. Setelah filtering, spektrum terlihat lebih bersih dan terfokus pada frekuensi utama suara, namun perbedaan pola dan puncak spektrum antar pembicara tetap terlihat jelas.

```python
def compute_fft(y, sr):
    Y = np.fft.rfft(y)
    freq = np.fft.rfftfreq(len(y), 1/sr)
    magnitude = np.abs(Y)
    return freq, magnitude

def bandpass_filter(signal, sr, lowcut=80, highcut=1100, order=4):
    nyquist = 0.5 * sr
    low = lowcut / nyquist
    high = highcut / nyquist

    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)

spectra_after  = {}

for folder in folders:
    mags_after  = []

    full_path = os.path.join(base_path, folder)

    for file in os.listdir(full_path):
        if file.endswith(".wav"):
            path = os.path.join(full_path, file)

            y, sr = librosa.load(path, sr=None)
            y = librosa.util.fix_length(y, size=sr * 3)

            # AFTER filter
            y_filt = bandpass_filter(y, sr)
            _, mag_after = compute_fft(y_filt, sr)
            mags_after.append(mag_after)

    spectra_after[folder]  = (freq, np.mean(mags_after, axis=0))
plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 2)
for folder, (freq, mag) in spectra_after.items():
    plt.plot(freq, mag, label=folder)
plt.title("Perbandingan Spektrum Rata-rata per Folder - After Filter")
plt.xlabel("Frekuensi (Hz)")
plt.ylabel("Magnitudo")
plt.legend()
plt.xlim(0, 5000)

plt.tight_layout()
plt.show()
```

Prediksi Suara

```python
import librosa
import numpy as np
import os
import json

DATASET_PATH = "/content/drive/Shareddrives/Pengolahan Sinyal Data/Dataset"
speaker_features = {}

for speaker in os.listdir(DATASET_PATH):
    speaker_path = os.path.join(DATASET_PATH, speaker)
    mfcc_list = []

    for file in os.listdir(speaker_path):
        if file.endswith(".wav"):
            audio_path = os.path.join(speaker_path, file)
            y, sr = librosa.load(audio_path, sr=16000)

            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=50)
            mfcc_mean = np.mean(mfcc, axis=1)
            mfcc_list.append(mfcc_mean.tolist())

    speaker_features[speaker] = np.mean(np.array(mfcc_list), axis=0).tolist()

with open("speaker_database.json", "w") as f:
    json.dump(speaker_features, f, indent=2)
```

Import librosa, numpy, os, dan json berfungsi sebagai tahap awal kebutuhan library yang akan digunakan untuk melakukan prediksi. DATASET_PATH berfungsi untuk menentukan lokasi dataset, dimana setiap folder merupakan satu pembicara dalam format .wav. Kita membuat dictionary speaker_features yang akan diisi oleh fitur dari setiap pembicara. Setelah itu, dilakukan loop untuk setiap folder pembicara dan menyusun path lengkap menuju folder pembicara, dan terakhir list mfcc akan disimpan dalam list kosong yang bernama mfcc_list. Setelah menentukan folder pembicara, kita lakukan loop pada setiap folder untuk mengambil file yang hanya berformat .wav. Hasil tersebut disimpan dalam audio_path untuk memudahkan ekstraksi fitur mfcc. Sebelum dilakukan proses ekstraksi fitur, dilakukan penyeragaman sampling rate menjadi 16.000 Hz yang bertujuan untuk konsistensi MFCC. Kemudian dilakukan Ekstraksi 50 fitur MFCC, dihitung rata-ratanya dan disimpan dalam mfcc_list. Agar file lebih mudah untuk diproses, dilakukan penyimpanan fitur kedalam file berformat .json.

```python
import json
import numpy as np
import librosa

db = json.load(open("speaker_database.json"))

def identify_speaker(mfcc_test):
    distances = {}
    for speaker, mfcc in db.items():
        dist = np.linalg.norm(np.array(mfcc_test) - np.array(mfcc))
        distances[speaker] = dist

    predicted = min(distances, key=distances.get)
    return predicted, distances


y, sr = librosa.load("/content/drive/Shareddrives/Pengolahan Sinyal Data/Dataset/Rama/Rama2.wav", sr=16000)

mfcc_test = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=50), axis=1)

predicted, distances = identify_speaker(mfcc_test)

print("Pembicara terdeteksi:", predicted)
print("\nJarak Euclidean ke tiap pembicara:")
for spk, dist in distances.items():
    print(f"  {spk}: {dist:.4f}")
```

Tahap awal yakni membaca file identitas dari setiap pembicara yang telah kita simpan dalam format .json. Kemudian membuat fungsi identify_speaker, didalamnya terdapat dictionary untuk menyimpan fitur suara uji dan identitas dari setiap pembicara. Loop semua pembicara dalam database dan menghitung jarak Euclidean antara vektor MFCC suara uji vektor MFCC rata-rata pembicara, hasilnya disimpan didalam distances. pada predicted, dipilih nilai terkecil yang akan menjadi hasil prediksi. Setelah membuat fungsi, kita menginput data yang akan diuji pada variable y,sr. Lalu file uji dihitung rata-rata pada mfcc_test dan akan dilakukan penerapang fungsi identify_speaker.
